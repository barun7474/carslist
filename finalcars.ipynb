```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

Http request

```python
url = 'https://www.cars.com/shopping/results/?dealer_id=&keyword=&list_price_max=&list_price_min=&makes[]=mercedes_benz&maximum_distance=30&mileage_max=&monthly_payment=&page_size=20&sort=best_match_desc&stock_type=cpo&year_max=&year_min=&zip='
```

```python
response = requests.get(url)
```

```python
response.status_code
```

Soup Object

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

Results

```python
results = soup.find_all('div', {'class':'vehicle-card'})
```

```python
results_length=len(results)
```

```python
results[1]
```

Targeting the necessary data

## Name
## Mileage
## rating
## review count
## price
## dealer name

```python
car_name =[]
for i in range(1,results_length):
    x=results[i].find('h2').get_text()
    car_name.append(x)
car_name
```

Rating

```python
rating = []

for i in range(1, results_length):
    result = results[i].find('span', {'class': 'sds-rating__count'})
    
    # Check if the 'result' is not None before trying to extract text
    if result is not None:
        y = result.get_text()
        rating.append(y)
    else:
        rating.append("No Rating Found")  # Or handle the absence of rating in any way you prefer

print(rating)
```

Review Count

```python
review=[]
for i in range(1,results_length):
  z=results[i].find('span',{'class':'test1 sds-rating__link sds-button-link'}).get_text().replace('(', '').replace(')', '')
  review.append(z)
review
```

Price

```python
price=[]
for i in range(1,results_length):
  a=results[i].find('span',{'class':'primary-price'}).get_text()
  price.append(a)
price
```

Mileage

```python
mileage=[]
for i in range(1,results_length):
  b=results[i].find('div',{'class':'mileage'}).get_text()
  mileage.append(b)
mileage
```

Dealer Name

```python
dealer_name=[]
for i in range(1,results_length):
  c=results[i].find('div',{'class':'dealer-name'}).get_text().replace('\n', '')
  dealer_name.append(c)
dealer_name
```

Converting data into dataframe

```python
data=pd.DataFrame()
data['car_name']=car_name
data['rating']=rating
data['review_count']=review
data['price']=price
data['mileage']=mileage
data['dealer_name']=dealer_name
data.head()
```
